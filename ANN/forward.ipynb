XOR Problem using Multi Layer Neural Network

```python
import numpy as np
```

Define XOR Dataset

```python
X=np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
X
```

Define Step Activation Function 

```python
def activation(z):
    return 1 if z>=0 else 0
```

```python
#Initialize Weights and Biases(Manually)
# Hidden layer weights
w1, w2 = 1, 1
w3, w4 = 1, 1
#bias
b1 = -0.5
b2 = -1.5

# Output layer weights
w5, w6 = 1, -2
b3 = -0.5
print("Weights intialized Succesfully")
```

Define Forward Propagation Function

```python
def forward_propagation(x1,x2):
    # Hidden layer
    h1 = w1*x1 + w2*x2 + b1
    h2 = w3*x1 + w4*x2 + b2
    a_h1 = activation(h1)
    a_h2 = activation(h2)
    #output layer calculation
    y=w5*a_h1 + w6*a_h2 + b3
    output=activation(y)
    return output
```

#Test Forward Propagation for All Xor Inputs

```python
print("Xor Results using ANN Forward Propagation")
for x in X:
    prediction=forward_propagation(x[0],x[1])
    print(f'Input:{x} Output: {prediction}')
```
