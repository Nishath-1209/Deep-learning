```python
import numpy as np
```

```python
X=np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
y=np.array([0,0,0,1])
X,y
```

```python
w=np.zeros(2)
b=0
learning_rate=0.1
epochs=10
print("Initial Weight:",w)
print("Initial Bias:",b)
```

```python
def step_function(z):
    return 1 if z>=0 else 0
    
```

```python
for epoch in range(epochs):
    print(f"\n Epoch {epoch+1}")
    for i in range(len(X)):
        # 1. Compute weighted sum
        z = np.dot(w,X[i]) + b
        # 2. Apply activation function
        y_pred = step_function(z)
        # 3. Compute error
        error = y[i] - y_pred 
        # 4. Update weight
        w = w + learning_rate * error * X[i]
        # 5. Update bias
        b = b + learning_rate * error
        print(f"Input:{X[i]}, Predicted:{y_pred}, Actual:{y[i]}, w:{w}, b:{b}")


```

```python
print("\n Final Weights",w)
print("\n Final Bias",b)
```

Test Model

```python
print("\n Testing AND Gate")
for i in range(len(X)):
    z=np.dot(w,X[i])+b
    prediction=step_function(z)
    print("Input",X[i],"Prediction:",prediction)
```

```python
import matplotlib.pyplot as plt
# Separate points by class
class0 = X[y == 0]
class1 = X[y == 1]

# Plot points
plt.scatter(class0[:, 0], class0[:, 1], marker='o')
plt.scatter(class1[:, 0], class1[:, 1], marker='x')

# Create decision boundary line
x_vals = np.linspace(-0.5, 1.5, 100)
y_vals = -(w[0] * x_vals + b) / w[1]

plt.plot(x_vals, y_vals)

# Labels and formatting
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Perceptron Decision Boundary (AND Gate)")
plt.grid(True)

plt.show()
```
