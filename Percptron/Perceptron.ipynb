Step-1
Prepare Training Data

```python
import numpy as np
```

```python
X=np.array([1,2,3,4]) #study hours
y=np.array([0,0,1,1]) #fail(0),pass(1)
X,y
```

```python
w=0.0
b=0.0
learning_rate=1
epochs=5
print("Initial weight:",w)
print("Initial Bias:",b)
```

Step 3:Define Activation Function(Step Function)

```python
def step_function(z):
    return 1 if z>=0 else 0
```

Training Loop

```python
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(X)):
        # 1. Compute weighted sum
        z = w * X[i] + b
        # 2. Apply activation function
        y_pred = step_function(z)
        # 3. Compute error
        error = y[i] - y_pred 
        # 4. Update weight
        w = w + learning_rate * error * X[i]
        # 5. Update bias
        b = b + learning_rate * error
        print(f"Input:{X[i]}, Predicted:{y_pred}, Actual:{y[i]}, w:{w}, b:{b}")
```

```python
print("\n Final Weight",w)
print("\n Final bias:",b)
```

Testing the Model

```python
test_value=3.1
z=w*test_value+b
prediction=step_function(z)
print(f"For Study Hours {test_value},  Prediction",prediction)
```

```python
import matplotlib.pyplot as plt
# --------- Plotting ---------
# Scatter plot of data
plt.scatter(X[y==0], y[y==0], color='red', label='Fail (0)')
plt.scatter(X[y==1], y[y==1], color='green', label='Pass (1)')
if w != 0:
    x_boundary = -b / w
    plt.axvline(x=x_boundary, color='blue', linestyle='--', label='Decision Boundary')
plt.xlabel("Study Hours")
plt.ylabel("Result")
plt.title("Perceptron Classification")
plt.legend()
plt.grid(True)
plt.show()
```
