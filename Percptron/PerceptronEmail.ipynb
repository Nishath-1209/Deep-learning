```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([
    [0, 0],
    [1, 0],
    [0, 1],
    [1, 1]
])
y = np.array([0, 1, 1, 1])
```

```python
w=np.zeros(2)
b=0
learning_rate=0.1
epochs=10
print("Initial Weight:",w)
print("Initial Bias:",b)
```

```python
def step_function(z):
    return 1 if z >= 0 else 0
```

```python
for epoch in range(epochs):
    for i in range(len(X)):
        z = np.dot(w, X[i]) + b
        y_pred = step_function(z)
        error = y[i] - y_pred
        w = w + learning_rate * error * X[i]
        b = b + learning_rate * error

print("Final Weights:", w)
print("Final Bias:", b)
```

```python
print("\nDecision Boundary Equation:")
print(f"{w[0]}*x1 + {w[1]}*x2 + {b} = 0")
class0 = X[y == 0]
class1 = X[y == 1]
plt.scatter(class0[:, 0], class0[:, 1], marker='o', label='Not Spam (0)')
plt.scatter(class1[:, 0], class1[:, 1], marker='x', label='Spam (1)')
# Create boundary line
x_vals = np.linspace(-0.5, 1.5, 100)
if w[1] != 0:
    y_vals = -(w[0] * x_vals + b) / w[1]
    plt.plot(x_vals, y_vals, label="Decision Boundary")
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)
plt.xlabel("Contains Suspicious Keywords (x1)")
plt.ylabel("Sender Unknown (x2)")
plt.title("Email Spam Detection using Perceptron")
plt.legend()
plt.grid(True)
plt.show()
```
